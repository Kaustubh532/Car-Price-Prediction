```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# manual - 1 Automatic - 0
# dealer - 1 Individual - 0
# Petrol - 1 Diesel - 2 CNG - 0
```

```python
df = pd.read_csv('C:/Users/Kaustubh Sawant/OneDrive/Desktop/machine learning/car data.csv')
df
```

```python
#actual age of car
df['Car_Age'] = 2025 - df['Year']
df
```

```python
# Target variable
y = df['Selling_Price']

# Features (excluding 'sell ing_price')
x = df.drop(['Selling_Price', "Car_Name"], axis=1)
```

```python
from sklearn.model_selection import train_test_split

# Split into training and testing (80-20)
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# initialize and train model 
lr_model = LinearRegression()
lr_model.fit(x_train, y_train)
#predict
y_pred_lr = lr_model.predict(x_test)
```

```python
#evaluation

r2 = r2_score(y_test, y_pred_lr)
mae = mean_absolute_error(y_test, y_pred_lr)
rmse = np.sqrt(mean_squared_error(y_test, y_pred_lr))


print("Linear Regression Model Performance:")
print("RÂ² Score:", r2)
print("MAE:", mae)
print("RMSE:", rmse)
```

```python
# random forest method
from sklearn.ensemble import RandomForestRegressor
# intialize the model
rf_model = RandomForestRegressor(n_estimators=100, random_state = 42)
#train
rf_model.fit(x_train, y_train) 
```

```python
y_pred_rf = rf_model.predict(x_test)
```

```python
r2_rf = r2_score(y_test, y_pred_rf)
mae_rf = mean_absolute_error(y_test, y_pred_rf)
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))

print("Random Forest Model Performance:")
print("RÂ² Score:", r2_rf)
print("MAE:", mae_rf)
print("RMSE:", rmse_rf)
```

```python
import joblib
joblib.dump(rf_model, 'Car_price_prediction.pkl')
```
